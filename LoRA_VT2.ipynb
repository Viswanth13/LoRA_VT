```python
from torchvision import transforms
from PIL import Image
import torch

# Define transformation pipeline
transform = transforms.Compose([
    transforms.Resize((512, 512)),  # Resize frames
    transforms.ToTensor(),  # Convert to tensor
])

# Load frames from dataset
def load_frames(dataset_path):
    images = []
    for anim_folder in sorted(os.listdir(dataset_path)):
        anim_path = os.path.join(dataset_path, anim_folder)
        if os.path.isdir(anim_path):
            for frame in sorted(os.listdir(anim_path)):
                frame_path = os.path.join(anim_path, frame)
                img = Image.open(frame_path).convert("RGB")
                images.append(transform(img))
    return torch.stack(images)

extract_path_new = "/content/animation_data/kaggle/working/animation_dataset"
dataset = load_frames(extract_path_new)
print(f"Loaded dataset shape: {dataset.shape}")
```

```python
import os
import torch
from diffusers import StableDiffusionPipeline, UNet2DConditionModel, AutoencoderKL
from transformers import  CLIPTextModel, CLIPTokenizer
from peft import LoraConfig, get_peft_model
from torch.utils.data import DataLoader, Dataset
from transformers import TrainingArguments, Trainer
from torchvision import transforms
from PIL import Image

# Define paths
model_id = "stabilityai/stable-diffusion-2-1"
output_dir = "./lora_finetuned"

# Load pretrained Stable Diffusion components
vae = AutoencoderKL.from_pretrained(model_id, subfolder="vae")
tokenizer = CLIPTokenizer.from_pretrained(model_id, subfolder="tokenizer")
text_encoder = CLIPTextModel.from_pretrained(model_id, subfolder="text_encoder")
unet = UNet2DConditionModel.from_pretrained(model_id, subfolder="unet")

# Apply LoRA to U-Net and Text Encoder
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["proj_in", "proj_out"],  # Apply LoRA to attention layers
    lora_dropout=0.05,
    bias="none",
)

unet = get_peft_model(unet, lora_config)
# text_encoder = get_peft_model(text_encoder, lora_config)

```

```python
class AnimationDataset(Dataset):
    def __init__(self, dataset_path, tokenizer, vae, caption="A travel-related animation"):
        self.dataset_path = dataset_path
        self.tokenizer = tokenizer
        self.vae = vae  # VAE used to compute latents
        self.caption = caption

        # Define a transformation to resize and normalize images.
        self.transform = transforms.Compose([
            transforms.Resize((512, 512)),
            transforms.ToTensor(),
            transforms.Normalize([0.5], [0.5]),  # Normalize between -1 and 1
        ])

        # Create a list of image paths from all animation folders.
        self.image_paths = []
        for anim_folder in sorted(os.listdir(dataset_path)):
            anim_path = os.path.join(dataset_path, anim_folder)
            if os.path.isdir(anim_path):
                for frame in sorted(os.listdir(anim_path)):
                    if frame.endswith(".png"):
                        self.image_paths.append(os.path.join(anim_path, frame))

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Load image and apply transforms.
        image_path = self.image_paths[idx]
        image = Image.open(image_path).convert("RGB")
        image = self.transform(image)

        # Compute latent representation using the VAE.
        # Add a batch dimension and move to the same device as the VAE.
        image_batch = image.unsqueeze(0).to(next(self.vae.parameters()).device)
        with torch.no_grad():
            latents = self.vae.encode(image_batch).latent_dist.sample()
        # Remove batch dimension for consistency.
        latents = latents.squeeze(0)

        # Tokenize the default caption.
        tokenized = self.tokenizer(self.caption, padding="max_length", max_length=77, return_tensors="pt")
        input_ids = tokenized.input_ids.squeeze(0)  # Remove batch dimension

        # Return a dictionary with the required keys.
        return {
            "pixel_values": latents,   # Latent representation of the image.
            "input_ids": input_ids,      # Tokenized default caption.
        }

```

```python
import torch
import torch.nn.functional as F
from torch.optim import AdamW
from torch.utils.data import DataLoader
from tqdm.auto import tqdm
from diffusers import UNet2DConditionModel, AutoencoderKL, DDPMScheduler
from transformers import CLIPTextModel, CLIPTokenizer
from peft import LoraConfig, get_peft_model

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load the pretrained components:
model_id = "stabilityai/stable-diffusion-2-1"

# Load U-Net, VAE, and noise scheduler from diffusers.
unet = UNet2DConditionModel.from_pretrained(model_id, subfolder="unet")
vae = AutoencoderKL.from_pretrained(model_id, subfolder="vae")
noise_scheduler = DDPMScheduler(num_train_timesteps=1000)

# Load the text encoder and tokenizer.
# Note: For Stable Diffusion 2.1, we use the OpenCLIP text encoder.
# text_encoder = CLIPTextModel.from_pretrained("openai/clip-vit-large-patch14")
# tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-large-patch14")



text_encoder = CLIPTextModel.from_pretrained(model_id, subfolder="text_encoder")
tokenizer = CLIPTokenizer.from_pretrained(model_id, subfolder="tokenizer")
```

```python
# Apply LoRA to the U-Net and text encoder.
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["proj_in", "proj_out"],  # Adjust these target modules if needed.
    lora_dropout=0.05,
    bias="none",
)
unet = get_peft_model(unet, lora_config)
# text_encoder = get_peft_model(text_encoder, lora_config)

# Move models to device.
unet.to(device)
text_encoder.to(device)
vae.to(device)
```

```python
from torch.utils.data import DataLoader
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)

# Set up the optimizer to update parameters from both the U-Net and the text encoder.
optimizer = AdamW(list(unet.parameters()) + list(text_encoder.parameters()), lr=1e-4)

# Training loop parameters.
num_epochs = 3

print("Starting training...")
for epoch in range(num_epochs):
    progress_bar = tqdm(dataloader, desc=f"Epoch {epoch+1}")
    for step, batch in enumerate(progress_bar):
        # Retrieve the latent image and tokenized caption.
        # "pixel_values" is the latent representation from your VAE.
        latents = batch["pixel_values"].to(device)   # Shape: [B, C, H, W]
        input_ids = batch["input_ids"].to(device)      # Shape: [B, seq_length]

        # Sample random noise.
        noise = torch.randn_like(latents)
        bsz = latents.shape[0]
        # Sample random timesteps for each image in the batch.
        timesteps = torch.randint(0, noise_scheduler.num_train_timesteps, (bsz,), device=device).long()

        # Add noise to the latents.
        noisy_latents = noise_scheduler.add_noise(latents, noise, timesteps)

        # Get text conditioning by passing the tokenized caption through the text encoder.
        encoder_outputs = text_encoder(input_ids)
        encoder_hidden_states = encoder_outputs.last_hidden_state

        # Forward pass through U-Net: predict the noise residual.
        model_output = unet(noisy_latents, timesteps, encoder_hidden_states)
        noise_pred = model_output.sample

        # Compute the loss: mean-squared error between the predicted noise and the actual noise.
        loss = F.mse_loss(noise_pred, noise)

        # Backpropagation.
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        progress_bar.set_postfix(loss=loss.item())
    print(f"Epoch {epoch+1} completed.")
```

```python
# Save fine-tuned model
unet.save_pretrained(output_dir)
text_encoder.save_pretrained(output_dir)
vae.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)

print("Fine-tuning completed! Model saved at", output_dir)
```
